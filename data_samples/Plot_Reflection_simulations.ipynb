```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(r"path to the data", header=None)

df_cleaned = df.iloc[1:, 1:].astype(float)
plt.imshow(df_cleaned.values[::-1], aspect='auto', cmap='jet', vmin=0, vmax=1)
plt.colorbar()
plt.xlabel(rf'$\theta (\degree) $', size=16)
plt.ylabel(rf'$\lambda$ (nm)', size=16)

num_rows = df_cleaned.shape[0]
y_ticks = np.linspace(0, num_rows - 1, 5)
y_labels = np.linspace(800, 400, 5).astype(int)
plt.yticks(y_ticks, y_labels, size=14)

plt.xticks(size=14)

# plt.savefig('Refl.png', dpi=600)
plt.show()
```
